```python
import matplotlib.pyplot as plt
from math import log, factorial
n = list(range(1,21))
y1 = [1 for n in n]
y2 = [log(n) for n in n]
y3 = [n for n in n]
y4 = [n*log(n) for n in n]
y5 = [n**2 for n in n]
y6 = [2**n for n in n]
y7 = [factorial(n) for n in n]
y8 = [n**n for n in n]
```

```python
plt.ylim(0, 20)


plt.plot(n, y1, color = 'blue', ls = '--', marker = 'o', ms = '5')
plt.plot(n, y2, color = 'red', ls = '--', marker = 'o', ms = '5')
plt.plot(n, y3, color = 'green', ls = '--', marker = 'o', ms = '5')
plt.plot(n, y4, color = 'yellow', ls = '--', marker = 'o', ms = '5')
plt.plot(n, y5, color = 'orange', ls = '--', marker = 'o', ms = '5')
plt.plot(n, y6, color = 'purple', ls = '--', marker = 'o', ms = '5')
plt.plot(n, y7, color = 'pink', ls = '--', marker = 'o', ms = '5')
plt.plot(n, y8, color = 'black', ls = '--', marker = 'o', ms = '5')

plt.title("Big O Notation")
plt.xlabel("Elements")
plt.ylabel("Operations")
plt.legend(['O(1)','$log(n)$','$O(n)$', 'O(n log n)', 'O(n^2)', ]);
    
```
