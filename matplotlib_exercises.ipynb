3. Combine the figures you created in the last step into one large figure with 4 subplots.

```python
plt.subplot(2,2,1)
plt.plot(x,y1, color = 'b', ls = '--', marker = 'o', ms = '5');
plt.title("$\sqrt{x}$")

plt.subplot(2,2,2)
plt.plot(x,y2, color = 'b', ls = '--', marker = 'o', ms = '5');
plt.title("$y=X^3$")

plt.subplot(2,2,3)
plt.plot(x,y3, color = 'b', ls = '--', marker = 'o', ms = '5');
plt.title("$y=2^X$")

plt.subplot(2,2,4)
plt.plot(x,y4, color = 'b', ls = '--', marker = 'o', ms = '5');
plt.title("$y=\dfrac{1}{x+1}$")

plt.suptitle('Math Functions', fontsize = 16)

plt.tight_layout();
```
