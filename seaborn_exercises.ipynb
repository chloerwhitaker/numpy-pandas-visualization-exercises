4. Using the chipotle dataset from the previous exercise, create a bar chart that shows the 4 most popular items and the revenue produced by each.

```python
def get_db_url(db_name):
    from env import host,user,password
    return f'mysql+pymysql://{user}:{password}@{host}/{db_name}'
url = get_db_url('chipotle')
sql = '''
SELECT *
FROM orders
'''
chipotle = pd.read_sql(sql, url)
chipotle.head()
```

```python
#most_orders = chipotle.groupby("item_name").agg({"id":"count","item_price":"sum"})
#most_orders.head()
```

```python
#four_most_pop = most_orders.sort_values(by="id",ascending=False).head(4)
#four_most_pop
```

```python
chipotle.item_price = chipotle.item_price.replace('[\$,]', '', regex=True).astype(float)
```

```python
most_orders = chipotle.groupby("item_name").agg({"id":"count","item_price":"sum"})
most_orders
type(most_orders)
most_orders.sort_values(by="id",ascending=False).head()
```

```python
top_four_items = most_orders.sort_values(by="id",ascending=False).head(4)
top_four_items
```

```python
tfi = top_four_items.reset_index().sort_values(by='item_price', ascending = False)
tfi
```

```python

plt.figure(figsize = (8, 5))

sns.barplot(y='item_name', x='item_price' ,data=tfi)
sns.set(rc={"figure.figsize":(5, 5)})

plt.title('Most popular items and revenue')

plt.ylabel("") # to get remove the column_name
plt.xlabel("Gross Revenue")
plt.suptitle("Revenue from Top 4 selling Items")
```
